```python
from casadi import Opti, pi
 
# Create an Opti instance
opti = Opti()
 
# Declare optimization variables
x = opti.variable()
y = opti.variable()
 
# Define the objective function
objective = (x - 3)**2 + (y - 2)**2
 
# Add objective to the Opti instance
opti.minimize(objective)
 
# Define the constraint
constraint = x**2 + y**2 - 13
 
# Add constraint to the Opti instance
opti.subject_to(constraint == 0)
 
# Define initial guesses for the optimization variables
opti.set_initial(x, 1)
opti.set_initial(y, 1)
 
# Option for JIT compilation
opts = {"jit": True,
        "compiler": "shell",
        "jit_options": {"flags": ["-O3"],
                         "verbose": True}}
 
# Use IPOPT solver with JIT compilation enabled
opti.solver("ipopt", opts)
 
# Solve the optimization problem
solution = opti.solve()
 
# Retrieve the solution
x_solution = solution.value(x)
y_solution = solution.value(y)
 
print(f"Optimal x: {x_solution}")
print(f"Optimal y: {y_solution}")
```
